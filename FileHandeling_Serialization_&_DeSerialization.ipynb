# Serialization_&_Deserialization

```python
import json
d = {
    'name' :'Prabir',
    "age" : 38,
    'salary' : 10000
}
l = [1,3,4,5,6]
with open('sample.json','w') as f:
  json.dump(d,f,indent = 4,sort_keys= True)  # indent is for space  # and sort_keys is use for sort the key

with open('sample.json','r') as f:
  print(json.load(f))  # load function is use for deserialize and we need pass just file handler
```

```python
# serialization and Deserialization in tuple
import json
t = (1,2,3,4,5)
with open('sample.json','w')as f:
  json.dump(t,f)  # it will convert into list
```

```python
#  list in dictionary
d = {
    'name' :'Prabir',
    "age" : 38,
    'salary' : 10000,
    'l' : [1,2,3,4,5]
}

with open('sample.json','w')as f:
  json.dump(d,f)
```

serialization and deserialization custom object


```python
class Person:
  def __init__(self,name,age,salary):
    self.name = name
    self.age = age
    self.salary = salary

p = Person('Prabir',39,10000)
p.salary
```

```python
def Show_object(p):
  if isinstance(p,Person):
    return f"name->{p.name} age->{p.age} salary->{p.salary}'"

import json

with open('sample.json','w')as f:
  json.dump(p,f,default = Show_object)
  # json.dump(p.__dict__,f,indent = 4) # another way to print
```

```python
import json
with open('sample.json','r')as f:
  p = json.load(f)
  print(p)
```

Pickling is use to convert into bytes

Unpickling is inverse se of pickling

```python
class Person:
  def __init__(self,name,age,salary):
    self.name = name
    self.age = age
    self.salary = salary
  def show(self):
    return f"name->{self.name} age->{self.age} salary->{self.salary}'"

p = Person('Prabir',39,10000)


import pickle
with open('sample.pkl','wb') as f:
  pickle.dump(p,f)  # Serialization
```

```python
import pickle
with open('sample.pkl','rb') as f:
  p = pickle.load(f)  # Deserialization
  print(p.show())
```
