```python
# write a function that can return the track record of 2 teams against each other

def record(ipl, team1, team2):

    matches = ipl[
        ((ipl['Team1'] == team1) & (ipl['Team2'] == team2))
        | ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))
    ]
    return matches

csk_rcb_record = record(ipl, 'Chennai Super Kings', 'Royal Challengers Bangalore')
csk_rcb_record['WinningTeam'].value_counts()
```

```python
# apply is use to perform some task

# apply function on series
temp = pd.Series([10,20,30,40,50])

def sigmoid(value):
  return 1/1+np.exp(-value)

temp.apply(sigmoid)

temp.apply(lambda x:x*10)


# apply function on DataFrame [use row by row]
def Eucledian_distance(row):
  pt_A = row['1st_point']
  pt_B = row['2nd_point']
  return ((pt_A[0]-pt_B[0])**2 + (pt_A[1]-pt_B[1])**2)**0.5


points_df = pd.DataFrame(
    {
        '1st_point':[(0,3),(2,3),(4,7),(3,9)],
        '2nd_point':[(11,3),(20,3),(4,17),(3,0)]
    }
)

points_df


points_df.apply(Eucledian_distance,axis=1)  # (axis = 1) because of row
```

```python
match_stadium = pd.read_csv('/content/matches.csv')
match_ball = pd.read_csv('/content/deliveries.csv')

temp_df = match_ball.merge(match_stadium,how='inner',left_on='match_id',right_on='id')

six_df = temp_df[temp_df['batsman_runs']==6]

num_six = six_df.groupby('venue')['venue'].count()

num_match = match_stadium['venue'].value_counts()

(num_six/num_match).sort_values(ascending=False)
```

```python
match_stadium
match_ball

temp_df = match_ball.merge(match_stadium,how='inner',left_on='match_id',right_on='id')

#  count sixes
sixes = temp_df[temp_df['batsman_runs']==6]

# number of sixes per stadium using groupby
sixes_per_stadium = sixes.groupby('venue')['venue'].count()

#  single matches per stadium
num_matches = match_stadium['venue'].value_counts()

(sixes_per_stadium/num_matches).sort_values(ascending=False).head(3)
```
